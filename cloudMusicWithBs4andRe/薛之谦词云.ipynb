```python
import numpy
import pandas
import os
from PIL import Image
import wordcloud
```

```python
geci_file = open('./LRC/薛之谦.txt')
lrczonghe = ''
with open('./LRC/dealXueZhiqian.txt','a+') as fp:
    for line in geci_file.readlines():
        lrczonghe +=line[10:].strip()
    fp.write(lrczonghe)
geci_file.seek(0)
geci_file.close()
```

```python
text = open('./LRC/dealXueZhiqian.txt').read()
```

```python
print(text)
```

```python
import jieba.analyse
```

```python
newtext = jieba.cut(text,cut_all=False)
```

```python
strtext = ' '.join(newtext)
```

```python
newstrtext = ' '.join(strtext.split('薛之谦'))
```

```python
newstrtext
```

```python
file_path = os.path.dirname('.')
```

```python
mask = numpy.array(Image.open(os.path.join(file_path,'xuezhiqian.jpg')))
```

```python
lrcwordCloud = wordcloud.WordCloud(font_path='./楷体_GB2312.ttf',mask=mask,max_words=200,max_font_size=40,background_color='white',stopwords=wordcloud.STOPWORDS.add("said"))
```

```python
lrcwordCloud.generate(newstrtext)
```

```python
lrcwordCloud.to_file(os.path.join(file_path,'xuezhiqianwordcloud.jpg'))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(lrcwordCloud)
plt.axis('off')
plt.show()
```
